```python
import requests
import datetime as dt
import pandas as pd
import tqdm
```

```python
station_id = {
    'LFPB': 'IGONES2', 
    'LTAC': 'IANKARA40',
    'GMMC': 'ICASABLA29',
    'VABB': 'IMUMBAI13',
    'RPLL': 'IPARAA8',
    'KATT': 'KTXAUSTI1253',
    'KTPA': 'KFLTAMPA399' 
}

columns = [
    'obsTimeUtc',
    'tempHigh',
    'tempLow',
    'tempAvg',
    'windspeedHigh',
    'precipRate',
    'precipTotal',
    'humidityHigh',
    'humidityLow',
    'humidityAvg'
]
```

```python
start = '20201115'
end = '20201231'
stationId = 'IGONES2'

df = pd.DataFrame(columns=columns)
for date in tqdm.notebook.tqdm(pd.date_range(start, end)):
    url = 'https://api.weather.com/v2/pws/history/daily?'
    params = [
        f'stationId={stationId}',
        'format=json',
        'units=m',
        'date={}'.format(date.strftime('%Y%m%d')),
        'apiKey=ce4df070611549708df0706115d970ee'
    ]

    url += '&'.join(params)
    r = requests.get(url)
    if r.ok:
        data = r.json()

        for observation in data['observations']:
            row = {}
            row['obsTimeUtc'] = pd.to_datetime(observation['obsTimeUtc']).date()
            row['tempHigh'] = observation['metric']['tempHigh']
            row['tempLow'] = observation['metric']['tempLow']
            row['tempAvg'] = observation['metric']['tempAvg']
            row['windspeedHigh'] = observation['metric']['windspeedHigh']
            row['precipRate'] = observation['metric']['precipRate']
            row['precipTotal'] = observation['metric']['precipTotal']
            row['humidityHigh'] = observation['humidityHigh']
            row['humidityLow'] = observation['humidityLow']
            row['humidityAvg'] = observation['humidityAvg']
            
            df = df.append(row, ignore_index=True)
```

```python
def item_generator(json_input, lookup_key):
    if isinstance(json_input, dict):
        for k, v in json_input.items():
            if k == lookup_key:
                yield lookup_key, v
            else:
                yield from item_generator(v, lookup_key)
    elif isinstance(json_input, list):
        for item in json_input:
            yield from item_generator(item, lookup_key)
            
list(item_generator(data, 'tempAvg'))
```
